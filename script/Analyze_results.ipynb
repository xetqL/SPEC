```python
import os
```

```python
files = list(filter(lambda f: 'Unloading' in f, os.listdir('.')))
results = []
alphas  = [0.03, 0.05, 0.1, 0.15, 0.18]
for f in files:
    with open(f, 'r') as of:
        data = of.read().split('\n')[-2]
        results.append((float(data.split(':')[-1])))
```

```python
files = list(filter(lambda f: 'Standard' in f, os.listdir('.')))
results_std = []
for f in files:
    with open(f, 'r') as of:
        data = of.read().split('\n')[-2]
        results_std.append((float(data.split(':')[-1])))
```

```python
results.sort()
results.reverse()
```

```python
import matplotlib.pyplot as plt
```

```python

plt.plot(alphas, results, label='Unloading model')
plt.plot(alphas, [results_std]*len(results), label='Zhai et al. 2018')

plt.xlabel('alpha')
plt.ylabel('time [s]', rotation='vertical')
plt.title('Time comparison b/w Zhai et al. (2018) vs. Unloading model w.r.t alpha values')
plt.legend()
```
